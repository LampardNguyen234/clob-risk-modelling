# Data Architecture and Market Characterization

## Executive Summary

This notebook establishes the foundational understanding of the cryptocurrency derivatives risk framework by exploring available data and developing robust statistical measures of volatility and liquidity. These metrics will drive the Open Interest (OI) limit and Price Band mechanisms developed in subsequent notebooks.

The analysis centers on a fundamental principle: **the cost to manipulate spot markets must exceed the potential profit from derivative positions**. By quantifying both volatility (measuring how much prices move) and liquidity (measuring how much capital is required to move prices), we can operationalize this principle into concrete risk parameters that protect the vault from insolvency while maintaining competitive market conditions.

## Section 1: Introduction and Risk Framework

### The Unique Challenges of Crypto Derivatives Risk Management

Cryptocurrency derivatives markets face distinct challenges compared to traditional financial markets. The core problem addressed by this framework is protecting the Insurance Fund (commonly called the "Vault") from insolvency caused by two primary threats: market manipulation attacks and extreme volatility events. Unlike traditional markets with deep institutional liquidity and regulatory oversight, crypto markets—particularly for smaller-cap tokens—can exhibit thin order books that make them vulnerable to deliberate price manipulation.

The framework addresses these threats through two critical mechanisms working in concert. First, **Open Interest Solvency Modeling** ensures that the economic cost of manipulating the underlying spot price exceeds any potential profit an attacker could extract from derivative positions. This creates a defensive moat where rational actors find attacks unprofitable. Second, **Dynamic Price Band Mechanisms** detect and reject erroneous or manipulative trades in real-time while preserving legitimate price discovery, preventing the vault from taking on positions at artificially inflated or deflated prices.

### The Insolvency Threat Model: The JellyJelly Case Study

The March 2025 JellyJelly attack on Hyperliquid provides a concrete illustration of how inadequate risk controls can lead to vault insolvency. Understanding this attack's mechanics is essential for appreciating why the framework developed in these notebooks is necessary.

**Attack Preparation**: The attackers began by funding three fresh wallets with approximately $7.2 million in capital. They accumulated spot JELLY tokens on Uniswap v3 while simultaneously providing liquidity to the JELLY/USDC pool. This dual strategy gave them both a large spot position and control over the liquidity depth of the pool. Critically, they also built significant Open Interest on Hyperliquid's JELLY perpetual market by taking net-neutral positions—one wallet held a large short position while two others held offsetting long positions. This neutrality prevented suspicion during the accumulation phase.

**The Manipulation**: Once positioned, the attackers executed a coordinated pump. They aggressively bought JELLY on Uniswap while simultaneously withdrawing their liquidity provision, creating a thin order book. This combination moved JELLY's price up approximately 500% in one hour. Because Hyperliquid's Index Price for JELLY relied heavily (or exclusively) on the Uniswap pool as its price source, the Index tracked the manipulated price upward. Consequently, the Mark Price increased proportionally.

**Vault Exploitation**: As Mark Price rose, short positions on the perpetual market showed massive unrealized losses and began triggering liquidations. The Hyperliquid vault (HLP), acting as the counterparty to these positions, absorbed the short liquidations and effectively inherited large short exposure at lower entry prices. Meanwhile, the attackers' long positions showed massive unrealized profits. The attackers then closed their long positions, extracting approximately $8 million in realized PnL from the vault. Finally, they dumped JELLY back onto the Uniswap pool, crashing the price and causing the oracle to revert to lower levels.

**The Aftermath**: The attackers successfully withdrew approximately $6 million before Hyperliquid could intervene. The vault was left holding losing short positions acquired through forced liquidations, resulting in substantial losses. This attack demonstrates the critical failure mode: when the cost to manipulate the Index is less than the potential profit from derivative positions, rational economic actors have strong incentives to attempt manipulation.

### The Fundamental Defense Equation

The JellyJelly attack crystallizes the core principle underlying this entire framework:

$$\text{Cost to Manipulate Spot} > \text{Potential Profit from Derivative}$$

If this inequality holds, the market is economically secure against rational attackers. Breaking down each side of this inequality clarifies the framework's design:

The **cost to manipulate** is primarily determined by spot market depth and liquidity. Moving a price from $P_0$ to $P_1$ requires absorbing all liquidity at intermediate price levels. In a deep, liquid market like BTC spot on Binance, moving the price even 1% might require hundreds of millions of dollars due to thick order books. In contrast, moving a thin DEX pool by 50% might cost only a few hundred thousand dollars. The Amihud Illiquidity Ratio, developed in Section 7 of this notebook, provides a quantitative measure of this cost.

The **potential profit from the derivative** is primarily determined by the allowed Open Interest and leverage. If an attacker can build a large leveraged long position, a manipulated price increase converts directly to realized profit upon closing the position. The key insight is that the exchange operator controls this side of the equation through OI limits. By restricting how large a position can be accumulated, the operator limits potential profit and makes attacks uneconomical.

The framework developed across these three notebooks operationalizes this inequality. Notebook 01 (this notebook) quantifies liquidity through ILLIQ metrics. Notebook 03 uses these liquidity metrics to derive OI limits that ensure the inequality holds. Notebook 02 provides real-time detection mechanisms to identify when manipulation attempts may be underway, enabling dynamic responses.

## Section 2: Data Loading and Validation

### Data Sources and Structure

The analysis relies on three data sources, each providing distinct information about price formation and trading activity:

**Mark Price Data** (`data/mark_price/`) contains 1-minute OHLC (Open, High, Low, Close) data representing the fair value price used for calculating Unrealized PnL and triggering liquidations. This is the most critical price for risk management, as it directly determines when positions are liquidated and how much collateral is required. Mark Price time series allow us to understand the volatility that traders and the vault actually experience.

**Index Price Data** (`data/index_price/`) contains 1-minute OHLC data representing the composite spot price from external exchanges. This serves as the anchor to external market reality and reveals whether the Mark Price is following broader market movements or deviating due to venue-specific issues. Comparing Index to Mark helps identify potential oracle manipulation or data quality problems.

**Last Price Data** (`data/klines/`) contains 1-minute OHLCV (Open, High, Low, Close, Volume) data for actual trades executed on the venue. Critically, this is the only data source that includes volume information. Volume data is essential for liquidity analysis, as it enables calculation of the Amihud Illiquidity Ratio and other metrics that relate price movement to trading activity. The klines dataset also includes additional fields such as quote volume, number of trades, and taker buy volumes that could be used for order flow analysis in Notebook 02.

### Data Constraints and Capabilities

The framework operates under a significant constraint: we have only 1-minute OHLCV data, not high-frequency tick data or order book snapshots. This limitation shapes what analysis is feasible.

**What We Cannot Do**: We cannot directly measure order book depth at specific price levels, which would require order book snapshots showing bid and ask quantities at each price point. We cannot detect millisecond-level spoofing or quote stuffing, which requires tick-by-tick order data. We cannot calculate precise slippage from order book data, which would require reconstructing the full limit order book and simulating order execution. These limitations mean we cannot build a microstructure-focused risk model that responds to sub-second market dynamics.

**What We Can Do**: We can construct robust statistical proxies for volatility using range-based estimators (Parkinson, Garman-Klass) that have been validated over decades of academic research. We can measure effective liquidity through the Amihud Illiquidity Ratio, which relates price impact to trading volume and has been shown to correlate strongly with true market depth. We can detect macroscopic solvency threats such as sustained illiquidity, extreme volatility regimes, and flash crashes that occur over minute-to-hour timescales. Importantly, the threats we care most about-large-scale manipulation attacks like JellyJelly—operate on these longer timescales and leave clear signatures in 1-minute data.


```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats.mstats import winsorize
from sklearn.linear_model import LinearRegression
import warnings
from pathlib import Path

# Configure display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
pd.set_option('display.float_format', lambda x: f'{x:.6f}')
warnings.filterwarnings('ignore')

# Set visualization style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('husl')

# Define constants
SYMBOLS = ['BTCUSDT', 'ETHUSDT', "SUIUSDT", "DOGEUSDT", "AXSUSDT"]
DATA_DIR = Path('../../data')
MARK_PRICE_DIR = DATA_DIR / 'mark_price'
INDEX_PRICE_DIR = DATA_DIR / 'index_price'
KLINES_DIR = DATA_DIR / 'klines'

def calculate_subplot_grid(n_plots):
    """
    Calculate optimal grid layout (nrows, ncols) for n_plots.
    Returns a grid that minimizes empty subplots while keeping aspect ratio reasonable.
    """
    if n_plots == 0:
        return 1, 1
    elif n_plots == 1:
        return 1, 1
    elif n_plots == 2:
        return 1, 2
    elif n_plots <= 4:
        return 2, 2
    elif n_plots <= 6:
        return 2, 3
    elif n_plots <= 9:
        return 3, 3
    elif n_plots <= 12:
        return 3, 4
    else:
        # For larger numbers, aim for roughly square grid
        ncols = int(np.ceil(np.sqrt(n_plots)))
        nrows = int(np.ceil(n_plots / ncols))
        return nrows, ncols

print("Environment initialized successfully")
print(f"Analyzing {len(SYMBOLS)} symbols: {', '.join(SYMBOLS)}")
```

### Data Loading Functions

```python
def load_mark_price(symbol):
    """Load mark price OHLC data for a symbol."""
    filepath = MARK_PRICE_DIR / f"{symbol}_1m.csv"
    df = pd.read_csv(filepath)
    df['open_time'] = pd.to_datetime(df['open_time'])
    df.set_index('open_time', inplace=True)
    df['symbol'] = symbol
    return df

def load_index_price(symbol):
    """Load index price OHLC data for a symbol."""
    filepath = INDEX_PRICE_DIR / f"{symbol}_1m.csv"
    df = pd.read_csv(filepath)
    df['open_time'] = pd.to_datetime(df['open_time'])
    df.set_index('open_time', inplace=True)
    df['symbol'] = symbol
    return df

def load_klines(symbol):
    """Load klines (last price) OHLCV data for a symbol."""
    filepath = KLINES_DIR / f"{symbol}_1m.csv"
    df = pd.read_csv(filepath)
    df['open_time'] = pd.to_datetime(df['open_time'])
    df.set_index('open_time', inplace=True)
    df['symbol'] = symbol
    # Keep relevant columns
    cols_to_keep = ['open', 'high', 'low', 'close', 'volume', 'quote_volume', 'trades', 
                    'taker_buy_base', 'taker_buy_quote', 'symbol']
    df = df[[col for col in cols_to_keep if col in df.columns]]
    return df

def load_all_data(symbols):
    """Load all three data sources for all symbols."""
    mark_data = {}
    index_data = {}
    klines_data = {}
    
    for symbol in symbols:
        try:
            mark_data[symbol] = load_mark_price(symbol)
            index_data[symbol] = load_index_price(symbol)
            klines_data[symbol] = load_klines(symbol)
            print(f"✓ Loaded {symbol}: {len(klines_data[symbol])} records")
        except FileNotFoundError as e:
            print(f"✗ Missing data for {symbol}: {e}")
    
    return mark_data, index_data, klines_data

# Load all data
mark_prices, index_prices, klines = load_all_data(SYMBOLS)
```

### Data Quality Assessment

```python
from IPython.display import HTML, display

def assess_data_quality(data_dict, data_type):
    """Assess data quality for a given data source."""
    quality_report = []
    
    for symbol, df in data_dict.items():
        report = {
            'Symbol': symbol,
            'Data Type': data_type,
            'Total Records': len(df),
            'Start Date': df.index.min(),
            'End Date': df.index.max(),
            'Duration (days)': (df.index.max() - df.index.min()).days,
            'Missing Values': df.isnull().sum().sum(),
            'Missing %': f"{df.isnull().sum().sum() / (len(df) * len(df.columns)) * 100:.2f}%"
        }
        
        # Check for temporal gaps (missing minutes)
        expected_records = (df.index.max() - df.index.min()).total_seconds() / 60 + 1
        gap_pct = (1 - len(df) / expected_records) * 100
        report['Temporal Gaps %'] = f"{gap_pct:.2f}%"
        
        quality_report.append(report)
    
    return pd.DataFrame(quality_report)

# Generate quality reports

display(HTML('<h4 style="text-align: center; margin-top: 15px; margin-bottom: 5px; font-weight: bold;">Mark Price Data</h4>'))
mark_quality = assess_data_quality(mark_prices, 'Mark Price')
display(mark_quality)

display(HTML('<h4 style="text-align: center; margin-top: 15px; margin-bottom: 5px; font-weight: bold;">Index Price Data</h4>'))
index_quality = assess_data_quality(index_prices, 'Index Price')
display(index_quality)

display(HTML('<h4 style="text-align: center; margin-top: 15px; margin-bottom: 5px; font-weight: bold;">Klines (Last Price) Data</h4>'))
klines_quality = assess_data_quality(klines, 'Klines')
display(klines_quality)
```

### Price Vector Relationships

Understanding how Last Price, Mark Price, and Index Price relate to each other reveals the market's structural characteristics. Large deviations between these prices can signal manipulation attempts, oracle failures, or extreme illiquidity.

```python
def analyze_price_relationships(symbol):
    """Analyze relationships between Last, Mark, and Index prices."""
    # Align data on common timestamps
    last = klines[symbol]['close'].rename('Last')
    mark = mark_prices[symbol]['close'].rename('Mark')
    index = index_prices[symbol]['close'].rename('Index')
    
    # Combine into single DataFrame
    prices = pd.concat([last, mark, index], axis=1).dropna()
    
    # Calculate deviations
    prices['Mark_Index_Basis_bps'] = (prices['Mark'] - prices['Index']) / prices['Index'] * 10000
    prices['Last_Mark_Deviation_bps'] = (prices['Last'] - prices['Mark']) / prices['Mark'] * 10000
    
    return prices

# Analyze price relationships for all symbols
price_relationships = {}
for symbol in SYMBOLS:
    if symbol in klines and symbol in mark_prices and symbol in index_prices:
        price_relationships[symbol] = analyze_price_relationships(symbol)

# Summary statistics
for symbol in SYMBOLS:
    if symbol in price_relationships:
        pr = price_relationships[symbol]
        print(f"\n{symbol}:")
        print(f"  Mark-Index Basis: μ={pr['Mark_Index_Basis_bps'].mean():.2f} bps, "
              f"σ={pr['Mark_Index_Basis_bps'].std():.2f} bps")
        print(f"  Last-Mark Deviation: μ={pr['Last_Mark_Deviation_bps'].mean():.2f} bps, "
              f"σ={pr['Last_Mark_Deviation_bps'].std():.2f} bps")
        print(f"  Correlation (Last-Mark): {pr['Last'].corr(pr['Mark']):.6f}")
        print(f"  Correlation (Mark-Index): {pr['Mark'].corr(pr['Index']):.6f}")
```

### Volume Characteristics

Volume analysis reveals trading activity patterns and helps identify periods of illiquidity. Assets with consistently low volume or frequent zero-volume candles require conservative risk parameters.

```python
# Volume statistics
volume_stats = []
for symbol in SYMBOLS:
    if symbol in klines:
        vol = klines[symbol]['volume'] * klines[symbol]['close'] # Volume in $
        stats_dict = {
            'Symbol': symbol,
            'Mean Volume': vol.mean(),
            'Median Volume': vol.median(),
            'P95 Volume': vol.quantile(0.95),
            'Max Volume': vol.max(),
            'Zero Volume %': (vol == 0).sum() / len(vol) * 100,
            'Daily Volume (avg)': vol.resample('D').sum().mean()
        }
        volume_stats.append(stats_dict)

volume_df = pd.DataFrame(volume_stats)

# Format the DataFrame for display
volume_df_display = volume_df.copy()
for col in ['Mean Volume', 'Median Volume', 'P95 Volume', 'Max Volume', 'Daily Volume (avg)']:
    volume_df_display[col] = volume_df_display[col].apply(lambda x: f'{x:,.2f}')
volume_df_display['Zero Volume %'] = volume_df_display['Zero Volume %'].apply(lambda x: f'{x:.4f}')

display(volume_df_display)
```

### Price Swing Analysis

Price swing analysis quantifies intraday volatility by measuring the percentage range of each 1-minute candle using log returns. The metric is defined as:

$$\text{Swing}\% = \ln\left(\frac{H_t}{L_t}\right) \times 100$$

where $H_t$ is the high price and $L_t$ is the low price for period $t$. 

Price swings reveal important characteristics about market microstructure. Assets with consistently small median swings but occasional extreme P99 swings exhibit "tail risk"—the typical behavior is stable, but rare events can be violent. Such assets require defensive Price Band calibration to avoid false triggers during normal operations while still catching genuine manipulation attempts. The timestamps of extreme swings provide ground truth data for validating volatility estimators and calibrating alert thresholds.

```python
def calculate_price_swings(symbol):
    """Calculate price swing percentage for each candle."""
    df = klines[symbol].copy()
    df['swing_pct'] = np.log(df['high'] / df['low']) * 100
    return df

# Calculate swings for all symbols
swing_data = {}
for symbol in SYMBOLS:
    if symbol in klines:
        swing_data[symbol] = calculate_price_swings(symbol)

# Generate summary statistics table
swing_summary = []
for symbol in SYMBOLS:
    if symbol in swing_data:
        swings = swing_data[symbol]['swing_pct']
        summary = {
            'Symbol': symbol,
            'Min': swings.min(),
            'Mean': swings.mean(),
            'Median': swings.median(),
            'Max': swings.max(),
            'P95': swings.quantile(0.95),
            'P99': swings.quantile(0.99),
            'P99/Median Ratio': swings.quantile(0.99) / swings.median()
        }
        swing_summary.append(summary)

swing_summary_df = pd.DataFrame(swing_summary)
print("\n=== PRICE SWING SUMMARY STATISTICS ===")
display(swing_summary_df.style.format({
    'Min': '{:.4f}',
    'Mean': '{:.4f}',
    'Median': '{:.4f}',
    'Max': '{:.2f}',
    'P95': '{:.4f}',
    'P99': '{:.4f}',
    'P99/Median Ratio': '{:.2f}x'
}))
```

The P99/Median Ratio is particularly revealing. Assets where this ratio exceeds 50x exhibit extreme tail risk: **their worst 1% of minutes experience price swings orders of magnitude larger than typical behavior**. Such asymmetry indicates vulnerability to manipulation or flash crashes and necessitates conservative risk parameters.

```python
# Visualize swing distributions
# Calculate grid layout dynamically based on number of symbols with data
symbols_with_data = [s for s in SYMBOLS if s in swing_data]
n_symbols = len(symbols_with_data)
nrows, ncols = calculate_subplot_grid(n_symbols)
fig, axes = plt.subplots(nrows, ncols, figsize=(6*ncols, 5*nrows))

# Handle both single subplot and multiple subplots
if n_symbols == 1:
    axes = [axes]
else:
    axes = axes.flatten() if n_symbols > 1 else [axes]

for idx, symbol in enumerate(symbols_with_data):
    ax = axes[idx]
    swings = swing_data[symbol]['swing_pct']
    
    # Plot histogram with KDE
    swings_clean = swings[swings < swings.quantile(0.999)]  # Remove extreme outliers for visualization
    ax.hist(swings_clean, bins=100, density=True, alpha=0.6, color='steelblue', edgecolor='none')
    
    # Add KDE
    from scipy.stats import gaussian_kde
    try:
        kde = gaussian_kde(swings_clean.dropna())
        x_range = np.linspace(swings_clean.min(), swings_clean.max(), 200)
        ax.plot(x_range, kde(x_range), 'r-', linewidth=2, label='KDE')
    except:
        pass
    
    # Add percentile markers
    p90 = swings.quantile(0.90)
    p95 = swings.quantile(0.95)
    p99 = swings.quantile(0.99)
    ax.axvline(p90, color='orange', linestyle='--', linewidth=1.5, label=f'P90: {p90:.4f}%')
    ax.axvline(p95, color='darkorange', linestyle='--', linewidth=1.5, label=f'P95: {p95:.4f}%')
    ax.axvline(p99, color='red', linestyle='--', linewidth=2, label=f'P99: {p99:.4f}%')
    
    ax.set_xlabel('Price Swing (%)')
    ax.set_ylabel('Density')
    ax.set_title(f'{symbol} - Price Swing Distribution')
    ax.legend(fontsize=8)
    ax.grid(True, alpha=0.3)

# Remove empty subplots if any
total_subplots = nrows * ncols
for idx in range(n_symbols, total_subplots):
    fig.delaxes(axes[idx])

plt.tight_layout()
plt.savefig('price_swing_distributions.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nFigure: Price swing distributions show the full range of intraday volatility.")
print("Heavy right tails indicate assets prone to sudden price spikes.")
print("Percentile markers (P90, P95, P99) help identify thresholds for alert calibration.")
```

## Section 4: Volatility Estimation Theory

### Why Standard Deviation of Close Prices Is Inadequate

Traditional volatility estimation uses the standard deviation of close-to-close returns, calculated as:

$$\sigma^2_{\text{Close}} = \frac{1}{n-1} \sum_{i=1}^{n} \left( \ln\left(\frac{C_i}{C_{i-1}}\right) \right)^2$$

where $C_i$ is the closing price at time $i$.

This estimator has a fundamental flaw for risk management: **it completely ignores the intraday path that price took during each period**.
- Consider a 1-minute candle where price drops 5% intraday (perhaps triggering liquidations or stop-losses) but closes exactly where it opened.
- The close-to-close return is zero, contributing zero to measured volatility, yet traders holding leveraged positions during that minute experienced significant drawdown and potential liquidation risk.

### The Parkinson Volatility Estimator

Parkinson Volatility is an estimator that uses only the High-Low range to estimate variance. The key insight is that if prices follow a Brownian motion (the standard assumption in quantitative finance), the expected squared range provides information about the underlying volatility.

For each period $t$, the Parkinson variance is:

$$\sigma^2_{\text{Parkinson},t} = \frac{1}{4 \ln(2)} \left( \ln\left(\frac{H_t}{L_t}\right) \right)^2$$

where $H_t$ is the high price, $L_t$ is the low price, and the constant $\frac{1}{4\ln(2)} \approx 0.361$ derives from the distribution of the range of Brownian motion. 

Parkinson Volatility is proven to be theoretically **5 times more efficient** than close-to-close estimation for a continuous random walk, meaning it requires only one-fifth of the data to achieve the same statistical precision.

The Parkinson estimator is particularly appropriate for the Price Band application developed in Notebook 02 because it explicitly captures "wicks"—the exact phenomenon we need to guard against. A manipulation attempt that briefly spikes the price will show up clearly in the H-L range even if the close quickly recovers. The estimator assumes continuous trading with no opening gaps, which holds well for cryptocurrency markets that trade 24/7/365.

The primary limitation of Parkinson is that it only uses High and Low, discarding information from Open and Close. For assets that trend consistently in one direction during a period, additional information can be extracted from whether the market closed higher or lower than it opened.

### The Garman-Klass Volatility Estimator

The Garman-Klass (GK) Volatility extends the range-based approach of Parkinson Volatility by incorporating Open and Close data. The Garman-Klass variance for period $t$ is:

$$\sigma^2_{\text{GK},t} = 0.5 \left( \ln\left(\frac{H_t}{L_t}\right) \right)^2 - (2\ln(2) - 1) \left( \ln\left(\frac{C_t}{O_t}\right) \right)^2$$

where the first term captures the range contribution (similar to Parkinson) and the second term adds information from the open-to-close movement. The constant $2\ln(2) - 1 \approx 0.386$ arises from minimizing the variance of the estimator.

GK is theoretically even more efficient than Parkinson (approximately **8 times more efficient than close-to-close**) and incorporates directional information showing whether the market closed higher or lower than it opened. This makes it superior for general Value-at-Risk calculations and risk reporting where understanding directional moves matters.

However, for Price Band construction specifically, Parkinson's simplicity and direct focus on extremes makes it more interpretable. The Price Band mechanism cares primarily about maximum deviation from fair value, not whether the market trended up or down.

## Section 5: Volatility Implementation

We now implement the three volatility estimators (Close-to-Close, Parkinson, Garman-Klass) and compare their behavior across assets and time periods.

```python
def calculate_close_to_close_volatility(df, window=60):
    """Calculate close-to-close volatility using log returns."""
    log_returns = np.log(df['close'] / df['close'].shift(1))
    volatility = log_returns.rolling(window=window).std()
    return volatility

def calculate_parkinson_volatility(df, window=60):
    """Calculate Parkinson volatility estimator."""
    # Parkinson variance for each period
    hl_ratio = np.log(df['high'] / df['low'])
    parkinson_var = (1 / (4 * np.log(2))) * (hl_ratio ** 2)
    # Rolling average of variance, then take square root
    volatility = np.sqrt(parkinson_var.rolling(window=window).mean())
    return volatility

def calculate_garman_klass_volatility(df, window=60):
    """Calculate Garman-Klass volatility estimator."""
    hl_ratio = np.log(df['high'] / df['low'])
    co_ratio = np.log(df['close'] / df['open'])
    
    # Garman-Klass variance for each period
    gk_var = 0.5 * (hl_ratio ** 2) - (2 * np.log(2) - 1) * (co_ratio ** 2)
    # Rolling average of variance, then take square root
    volatility = np.sqrt(gk_var.rolling(window=window).mean())
    return volatility

def annualize_volatility(vol_minute):
    """Convert per-minute volatility to annualized volatility."""
    minutes_per_year = 365.25 * 24 * 60
    return vol_minute * np.sqrt(minutes_per_year)

# Calculate volatilities for all symbols
volatility_results = {}
window_size = 60  # 60-minute window for balanced responsiveness

for symbol in SYMBOLS:
    if symbol in klines:
        df = klines[symbol].copy()
        
        # Calculate all three estimators
        df['vol_close'] = calculate_close_to_close_volatility(df, window=window_size)
        df['vol_parkinson'] = calculate_parkinson_volatility(df, window=window_size)
        df['vol_gk'] = calculate_garman_klass_volatility(df, window=window_size)
        
        # Annualize
        df['vol_close_annual'] = annualize_volatility(df['vol_close']) * 100  # Convert to percentage
        df['vol_parkinson_annual'] = annualize_volatility(df['vol_parkinson']) * 100
        df['vol_gk_annual'] = annualize_volatility(df['vol_gk']) * 100
        
        volatility_results[symbol] = df

print("✓ Volatility calculations completed for all symbols")
```

## Section 6: Liquidity Theory

While volatility measures how much prices move, liquidity measures how much capital is required to move prices. This distinction is critical: **a highly volatile but liquid asset (like Bitcoin during a news event) poses less manipulation risk than a stable but illiquid asset (like a small-cap token where $100k can move the price 10%)**.

The new recommended metrics are:
1. **Corwin-Schultz Spread (2012)** - Primary measure for bid-ask costs
2. **Kyle's Lambda (1985/2016)** - Secondary measure for market impact
3. **Amihud ILLIQ (2002)** - DEPRECATED but kept for comparison

### The Corwin-Schultz Spread Estimator (2012) ⭐ RECOMMENDED

The Corwin-Schultz (CS) Spread Estimator is specifically designed for OHLC data that has become the industry standard.

**Key Insight**: Daily high prices are almost always buy orders (executed at ask), and daily low prices are almost always sell orders (executed at bid). The high-to-low ratio reflects both variance and spread.

**For each period $t$:**

$$\beta_t = \left(\ln\frac{H_t}{L_t}\right)^2$$

$$\gamma_t = \left(\ln\frac{H_{2t}}{L_{2t}}\right)^2$$

where $H_{2t}$ and $L_{2t}$ are the maximum high and minimum low over a 2-period window. By comparing 1-period vs 2-period ratios (variance doubles but spread remains constant), we can isolate the spread component.

**Component $\alpha$ (spread-related):**

$$\alpha_t = \frac{\sqrt{2\beta_t} - \sqrt{\beta_t}}{3 - 2\sqrt{2}} - \sqrt{\frac{\gamma_t}{3 - 2\sqrt{2}}}$$

**Estimated spread:**

$$S_t = \frac{2(e^{\alpha_t} - 1)}{1 + e^{\alpha_t}}$$

**Interpretation**:
- $S = 0.001$ (0.1%) = Very tight spread, highly liquid (Tier 1: BTC, ETH)
- $S = 0.005$ (0.5%) = Moderate spread (Tier 2: established altcoins)
- $S = 0.01$ (1.0%) = Wide spread, illiquid (Tier 3: emerging tokens)
- $S > 0.02$ (2.0%) = Very illiquid, high manipulation risk (Tier 4)

**Advantages over Amihud**:
- ✅ Best performer for time-series variation in cryptocurrency markets (Hautsch et al. 2022)
- ✅ No 24/7 trading mismatch issues
- ✅ Captures actual trading costs (bid-ask crossing)
- ✅ More stable values (avoids extremely small numbers)
- ✅ Validated specifically for crypto markets

**Reference**: Corwin, S. A., & Schultz, P. (2012). "A Simple Way to Estimate Bid-Ask Spreads from Daily High and Low Prices." *The Journal of Finance*, 67(2), 719-760.

### Kyle's Lambda for Price Impact (1985/2016) ⭐ RECOMMENDED

While the Corwin-Schultz spread measures the fixed cost of crossing the bid-ask spread, large orders also experience **market impact** - the price movement caused by the order itself. Albert Kyle's seminal 1985 paper "Continuous Auctions and Insider Trading" and the 2016 extension by Kyle & Obizhaeva quantify this.

Kyle's Lambda ($\lambda$) represents the price impact coefficient:

$$\Delta P_t = \lambda \times \sqrt{Q_t} + \varepsilon_t$$

where $Q_t$ is the order size in dollars. The **square-root relationship** has been validated across many markets and is known as the "market impact invariance" principle.

**Estimation**: We estimate $\lambda$ from historical data using rolling regression:
- Calculate absolute price changes: $|\Delta P| = |H_t - L_t|$
- Calculate $\sqrt{\text{DollarVolume}}$
- Regress: $|\Delta P| \sim \lambda \times \sqrt{\text{DollarVolume}}$

**Interpretation**:
- $\lambda = 10^{-6}$: Very low impact (highly liquid)
- $\lambda = 10^{-5}$: Moderate impact
- $\lambda = 10^{-4}$: High impact (illiquid)
- $\lambda > 10^{-3}$: Extreme impact (manipulation risk)

**Combined Liquidity Cost for OI Limits**:

When liquidating a position of size $Q$ (USD), the total liquidity cost is:

$$\text{TotalCost} = \underbrace{S}_{\text{spread crossing}} + \underbrace{\lambda \sqrt{Q}}_{\text{market impact}}$$

This combined cost must remain below the maintenance margin to prevent bad debt. This is the foundation of the OI limit formula in Notebook 03.

**Reference**: Kyle, A. S., & Obizhaeva, A. A. (2016). "Market Microstructure Invariance." *Econometrica*, 84(4), 1345-1404.

### The Amihud Illiquidity Ratio (2002) ⚠️ DEPRECATED

**DEPRECATION NOTICE**: The Amihud ILLIQ ratio is kept for historical comparison but is **NOT RECOMMENDED** for production use in cryptocurrency markets. Use Corwin-Schultz spread + Kyle Lambda instead.

**Why Deprecated**:
1. ❌ Denominator-numerator mismatch (24/7 trading issue)
2. ❌ Values too small (scientific notation, hard to interpret)
3. ❌ Volume-dominated (doesn't capture true price impact)
4. ❌ No fixed-cost component (only captures variable costs)

See `docs/liquidity_estimation_alternatives.md` for detailed analysis.

---

**Original Theory** (Amihud 2002):

Yakov Amihud's paper "Illiquidity and Stock Returns" introduced ILLIQ as a measure of price impact per dollar traded:

$$\text{ILLIQ}_t = \frac{|R_t|}{\text{Volume}_t \times P_t}$$

where:
- $R_t$ is the absolute return (we use intraday range: $\frac{H_t - L_t}{L_t}$)
- $\text{Volume}_t$ is trading volume in base currency
- $P_t$ is the price
- Denominator = dollar volume

**Interpretation** (for reference only):
- $\text{ILLIQ} = 10^{-6}$: Very liquid (BTC, ETH)
- $\text{ILLIQ} = 10^{-5}$: Moderate liquidity
- $\text{ILLIQ} = 10^{-4}$: Illiquid
- $\text{ILLIQ} \geq 10^{-3}$: Extreme illiquidity

**Reference**: Amihud, Y. (2002). "Illiquidity and Stock Returns: Cross-Section and Time-Series Effects." *Journal of Financial Markets*, 5(1), 31-56.

**⚠️ DO NOT USE for new calculations**. Results included only for validation and comparison purposes.

### OI Capacity Preview

Using the OI limit formula preview from Section 6, we can estimate the Open Interest capacity for each asset. This is a simplified calculation; Notebook 03 will develop the full methodology including dynamic adjustments.

```python
# ============================================================================
# OI CAPACITY PREVIEW (Dual Methodology - Both Methods)
# ============================================================================

def estimate_oi_capacity_illiq(illiq_mean, maintenance_margin=0.05, alpha=0.4):
    """
    Estimate OI capacity using Amihud ILLIQ method.
    
    Method 2 formula: OI_limit = alpha * (MM / ILLIQ_24h)
    
    Returns OI limit as multiple of maintenance margin.
    """
    if illiq_mean <= 0 or np.isnan(illiq_mean):
        return np.nan
    
    oi_limit_ratio = alpha * (maintenance_margin / illiq_mean)
    return oi_limit_ratio


# Apply ILLIQ-based estimation (Method 2)
liquidity_summary_df['OI_Capacity_ILLIQ'] = liquidity_summary_df['ILLIQ_24h'].apply(
    estimate_oi_capacity_illiq
)

print("\n" + "=" * 80)
print("OI CAPACITY PREVIEW (Method 2: Amihud ILLIQ)")
print("=" * 80)
print("Formula: OI_limit = α × (MM / ILLIQ_24h)")
print("Parameters: α=0.4, MM=5%")
print("\nNote: This is a preview using Method 2 (ILLIQ)")
print("Notebook 03 will calculate OI limits using BOTH methods:")
print("  - Method 1: OI = α × ((MM - S) / λ)²")
print("  - Method 2: OI = α × (MM / ILLIQ)")
print("=" * 80 + "\n")

display_cols = [
    'Symbol',
    'CS_Spread_24h_Mean_%',
    'ILLIQ_24h',
    'OI_Capacity_ILLIQ',
    'Liquidity_Tier_CS'
]

display(liquidity_summary_df[display_cols].style.format({
    'CS_Spread_24h_Mean_%': '{:.4f}%',
    'ILLIQ_24h': '{:.2e}',
    'OI_Capacity_ILLIQ': '{:.2e}'
}))

print("\n" + "=" * 80)
print("This is a preliminary estimate using Method 2 (Amihud ILLIQ)")
print("")
print("Notebook 03 will implement BOTH methodologies:")
print("")
print("Method 1 (CS Spread + Kyle Lambda):")
print("  OI_limit = α × ((MM - S) / λ)²")
print("  Where: S = CS_spread, λ = kyle_lambda")
print("")
print("Method 2 (Amihud ILLIQ):")
print("  OI_limit = α × (MM / ILLIQ)")
print("")
print("Both will be calculated and compared side-by-side")
print("See docs/dual_methodology_approach.md for details")
print("=" * 80)
```

### Synthetic Liquidation Simulation

We simulate liquidation scenarios to estimate at what position size the vault would face insolvency risk. This analysis validates that the OI capacity estimates are reasonable.

```python
def simulate_liquidation_slippage(illiq_24h, position_value_usd, maintenance_margin=0.05):
    """
    Simulate expected slippage when liquidating a position using Amihud ILLIQ.
    
    Parameters:
    - illiq_24h: 24-hour ILLIQ ratio
    - position_value_usd: Dollar value of position to liquidate
    - maintenance_margin: Margin rate (e.g., 0.05 for 5%)
    
    Returns:
    - slippage_pct: Expected slippage as percentage
    - causes_bad_debt: Boolean indicating if slippage exceeds margin
    """
    slippage_pct = illiq_24h * position_value_usd
    causes_bad_debt = slippage_pct > maintenance_margin
    return slippage_pct, causes_bad_debt

# Test various position sizes
test_position_sizes = [100_000, 500_000, 1_000_000, 5_000_000, 10_000_000]  # USD

print("\n=== SYNTHETIC LIQUIDATION SIMULATION (Method 2: Amihud ILLIQ) ===")
print("Testing whether liquidation of various position sizes would cause bad debt\n")

for symbol in SYMBOLS:  # Test on representative assets
    if symbol in liquidity_summary_df['Symbol'].values:
        illiq_mean = liquidity_summary_df[liquidity_summary_df['Symbol'] == symbol]['ILLIQ_24h'].iloc[0]
        
        print(f"{symbol} (ILLIQ_24h = {illiq_mean:.2e}):")
        for pos_size in test_position_sizes:
            slippage, bad_debt = simulate_liquidation_slippage(illiq_mean, pos_size)
            status = "⚠️ BAD DEBT" if bad_debt else "✓ Safe"
            print(f"  ${pos_size:>10,} position: {slippage*100:.3f}% slippage — {status}")
        print()
```

The simulation reveals the critical threshold where liquidation slippage exceeds the maintenance margin buffer. Assets with low ILLIQ can safely support much larger positions before risking bad debt, while high-ILLIQ assets hit the insolvency threshold at much smaller position sizes. This validates the inverse relationship between ILLIQ and OI capacity.
